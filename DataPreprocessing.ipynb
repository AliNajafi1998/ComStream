# Preprocessing

```
extractor = URLExtract()

def clean_data(string):
    urls = set(extractor.gen_urls(string))
    for url in urls:
        string = string.replace(url, '')
    
    string =  re.sub(r'[^\w\s#\'@]','',string)
 
    x = []
    for i in string.split():
        token = i.lower()
        if (not '@' in token) and (not token in stop_words_set) and ("'" not in token):
            x.append(token)
 
 
    return " ".join(x)
```

```
df['text'] = df['text'].apply(clean_data)
```
