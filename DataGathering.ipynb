```
import os
import numpy as np
import pandas as pd
```

```
early_data_path = os.getcwd() + '/RawData/Early'
late_data_path =  os.getcwd() + '/RawData/Late'
```

```
def get_en_tweets(new_df) -> pd.DataFrame:
    new_df = new_df[new_df['lang'] == 'en']
    return new_df

def read_raw_data(path=early_data_path,start=0,end=18)-> None:
    df = pd.DataFrame()
    files_names = os.listdir(path)
    if len(files_names) > end:
        end = len(files_names)
    for file_name in files_names[start:end]:
        new_df = pd.read_csv(path+'/'+file_name)
        new_df = get_en_tweets(new_df)
        new_df = new_df[['user_id','status_id','text','created_at','verified','favourites_count','retweet_count']]
        df = df.append(new_df, ignore_index = True)
    return df

def write_data_to_csv(df:pd.DataFrame,dir_)->None:
    df.to_csv(dir_,index=False)
```

```
df_early = read_raw_data(path=early_data_path)
df_late = read_raw_data(path=late_data_path)
```
